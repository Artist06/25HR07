```python
import pandas as pd
from sklearn.metrics import (
    confusion_matrix, classification_report,
    f1_score, accuracy_score
)
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def show_evaluation_metrics(input_csv_path : str):
    input_file = input_csv_path
    df = pd.read_csv(input_file)
    df.drop(columns=["Unnamed: 0"], inplace=True)
    
    y_true = df["actual_label"]
    y_pred = df["predicted_label"]
    
    print(f"\nF1 Score: {f1_score(y_true, y_pred)}")
    print(f"Accuracy: {accuracy_score(y_true, y_pred)}")
    print("\nDetailed Classification Report:")
    print(classification_report(y_true, y_pred))
    
    cm = confusion_matrix(y_true, y_pred)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.title('Confusion Matrix')
    plt.show()
```

# LSTM

```python
show_evaluation_metrics('./model/content_lstm_predictions.csv')
```

# BiLSTM

```python
show_evaluation_metrics('./model/content_bilstm_predictions.csv')
```

# GRU

```python
show_evaluation_metrics('./model/content_gru_predictions.csv')
```

```python
show_evaluation_metrics('./model/content_bigru_predictions.csv')
```
