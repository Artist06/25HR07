```python
import mysql.connector
import pandas as pd
import pickle
import time
from urllib.parse import urlparse, urljoin
from bs4 import BeautifulSoup
```

```python
conn = mysql.connector.connect(
    host="localhost",
    user="root",
    password="srikant",
    database="phishing_websites"
)
        
cursor = conn.cursor()
```

```python
cursor.execute('select rec_id, url, website, result from ph')
query = cursor.fetchall()
```

```python
html_folder = './dataset/'
output_folder = './doms/'

for i in range(0, 80000, 10000):
    start_files = i
    end_files = i + 10000
    message = f"Calculating from {start_files} to {end_files}"
    print('*'*len(message))
    print(message)
    print('*'*len(message))
    d = [list(i) for i in query]
    start = time.time()
    beginning = start

    for i in range(start_files, end_files):
        with open(html_folder + d[i][2], 'r', encoding='utf-8') as html_doc:
            soup = BeautifulSoup(html_doc.read(), 'lxml')
            d[i][2] = soup
        
        if i%1000 == 0 and i-start_files:
            end = time.time()
            print(f"Current Progress: {i-start_files}/{end_files - start_files}, Time taken: {end - start:.2f}s")
            start = time.time()
            
    pickle.dump(d[start_files:end_files], open(output_folder+f"dom_data{int(start_files/10000)}.pkl", "wb+"))
    print(f"Total time taken: {time.time() - beginning:.2f}s")
```
